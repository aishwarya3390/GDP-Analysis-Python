```python
from bs4 import BeautifulSoup
import requests
```

```python
url ="https://cleartax.in/s/world-gdp-ranking-list"
page = requests.get(url)
page 
```

```python
soup = BeautifulSoup(page.text)
soup 
```

```python
table= soup.find("table")
table
```

```python
table= soup.find("tbody")
table
```

```python
rows = table.find_all("tr")
rows 
```

```python
rows = table.find_all("tr")
rows 
```

```python
gdp_data = []

for row in rows[1:]:
    cols = row.find_all(["td", "th"])
    row_data = [col.text.strip().replace('\xa0', ' ') for col in cols]
    gdp_data.append(row_data)
```

```python
gdp_data 
```

```python
print (gdp_data)
```

```python
import pandas as pd

df = pd.DataFrame(gdp_data, columns=[
    "Rank", "Country", "Continent", "GDP (USD Trillions)", "Per Capita GDP"
])
```

```python
df
```

```python
df.to_csv("World-GDP.csv", index=False)
print("File saved as 'country-gdp.csv'")
```
